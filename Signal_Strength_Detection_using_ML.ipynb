# **Pre-Processing the Data**

```python
signal_data.columns
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
le = LabelEncoder()
scaler = StandardScaler()
```

```python
signal_data['Network Type'] = le.fit_transform(signal_data['Network Type'])
```

```python
signal_data.head()
```

```python
signal_data.columns
```

```python
signal_data.drop(columns=['Locality'])
```

# ***Defining the Features and Target***

```python
X = signal_data.drop(columns=['Signal Strength', 'Locality'])  # Features
y = signal_data['Signal Strength']                 # Target
```

```python
X.shape
```

```python
y.shape
```

```python
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

# **Model Training**

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
# Training the  model

rfr_model = RandomForestRegressor(n_estimators=100, random_state=42)
rfr_model.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
# Predict and evaluate

y_pred = rfr_model.predict(X_test)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Absolute Error: {mae:.3f}")
print(f"\nMean Squared Error: {mse:.3f}")
print(f"\nR² Score: {r2:.3f}")
```

# **Predicted vs Actual Signal Strength Plot**

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x=y_test, y=y_pred, alpha=0.5)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], '--r', linewidth=2)
plt.xlabel('Actual Signal Strength (dBm)')
plt.ylabel('Predicted Signal Strength (dBm)')
plt.title('Actual vs Predicted Signal Strength')
plt.grid(True)
plt.show()
```

# **Feature Importance**

```python
# Get feature importances and labels

feature_names = X.columns
importances = rfr_model.feature_importances_
```

```python
# Plot
plt.figure(figsize=(8, 5))
sns.barplot(x=importances, y=feature_names)
plt.title('Feature Importances from Random Forest')
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.grid(True)
plt.show()
```
